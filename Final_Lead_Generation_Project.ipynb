```python
import pandas as pd
```

```python
df = pd.read_csv("Indian Unicorn startups 2023 updated.csv")
```

```python
print(df)
```

```python
df["Company"] = (
    df["Company"]
    .str.replace("^", "", regex=False)
    .str.replace("*", "", regex=False)
    .str.strip()
)

print(df.head(15))
```

```python
df = df[["Company", "Sector", "Location"]]
df.head()
```

```python
df["Website"] = (
    "https://www."
    + df["Company"]
    .str.lower()
    .str.replace(" ", "", regex=False)
    .str.replace("^", "", regex=False)
    .str.replace("*", "", regex=False)
    + ".com"
)

print(df.head())
```

```python
df.drop_duplicates(inplace=True)

print(df.shape)
```

```python
df["Generated Email"] = (
    df["Company"]
    .str.lower()
    .str.replace(" ", "", regex=False)
    .str.replace("^", "", regex=False)
    .str.replace("*", "", regex=False)
    + "@gmail.com"
)

print(df.head())
```

```python
df["LinkedIn"] = (
    "https://www.linkedin.com/search/results/all/?keywords="
    + df["Company"]
    .str.replace(" ", "%20")
)

print(df.head())
```
